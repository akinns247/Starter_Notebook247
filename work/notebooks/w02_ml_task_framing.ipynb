## 1. My lane as an ML task (type)

ML Task Type: Ranking

This is a ranking problem because the goal is to order website pages from highest to lowest priority for content refresh. Rather than simply deciding whether a page should or should not be refreshed, the model ranks all pages based on historical SEO and engagement data (such as CTR, average search position, traffic trends, and engagement rate). This allows content managers to focus first on the pages that are most likely to benefit from an update.

# Target or Proxy

There is no direct "needs refresh" label in the dataset, so I will use a proxy target.

A possible proxy is:

- Low Click-Through Rate (CTR)
- Declining traffic trend
- Poor average search position
- Low engagement rate

These indicators together suggest that a page may benefit from a content refresh.

Example target:

needs_refresh = 1 if
CTR is low AND trend_direction is "down" AND average_position is poor.

Otherwise:
needs_refresh = 0

# Success Metric

The success of this ML task can be measured using ranking quality.

Possible metrics include:

- Precision@K (How many of the top recommended pages truly needed refreshing)
- Recall
- F1 Score (if converted into classification)
- NDCG (Normalized Discounted Cumulative Gain) for ranking quality

The most useful metric is Precision@K because content teams usually update only a limited number of pages.
